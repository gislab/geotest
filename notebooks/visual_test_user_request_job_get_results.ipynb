```python
%matplotlib inline
from django.db.models import Q
import matplotlib.pyplot as plt
from coordmanager.models import UserRequestJob
plt.style.use('seaborn-whitegrid')
query_point = (551877.0, 1325102.2)
num_point = 500
```

```python
nearest = UserRequestJob(x=query_point[0], y=query_point[1], num_point=num_point, operation=UserRequestJob.NEAREST)
furthest = UserRequestJob(x=query_point[0], y=query_point[1], num_point=num_point, operation=UserRequestJob.FURTHEST)

for point in nearest.get_results():
    plt.plot(point[0], point[1], 'o', color='red')
for point in furthest.get_results():
    plt.plot(point[0], point[1], 'o', color='yellow')

plt.plot(query_point[0], query_point[1], 'x', color='blue')
plt.show()
```
